```python
le = LabelEncoder()

data1['education'] = le.fit_transform(data1['education'])
data1['experience'] = le.fit_transform(data1['experience'])
data1['industry'] = le.fit_transform(data1['industry'])
data1['skills'] = le.fit_transform(data1['skills'])
data1['jobtitle'] = le.fit_transform(data1['jobtitle'])
```

```python
data1['skills'].value_counts()
```

```python
data1.head()
```

```python
# normalization

def nor_func(i):
    x= (i-i.min())/(i.max()-i.min())
    return(x)
```

```python
data2= nor_func(data1)
```

```python
data2.head()
```

```python
x= data2.iloc[:,:-1]
y= data2.iloc[:,-1]
```

```python
x.shape,y.shape
```

```python
from sklearn.cluster import KMeans
```

```python
data3 = []
for value in range(1,11):
    km = KMeans(n_clusters=value,init="random", n_init=10, max_iter=300)
    km.fit(x, y)
    data3.append(km.inertia_)
```

```python
data3
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(1, 11), data3)
plt.xticks(range(1,11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```
